(1) Pie chart

```python
df_for_pie = datasetEE.groupby('Country', axis=0).sum()
df_for_pie
```

```python
explode_list = [0.1, 0, 0.1, 0, 0, 0.1, 0, 0, 0.1, 0]

df_for_pie['Total'].plot(kind='pie', figsize=(15, 6), autopct='%1.1f%%', startangle=90, labels=None, pctdistance=1.2, explode=explode_list)

plt.title('Immigration to Canada from Eastern Europe for 1980-2013', y=1.12)
plt.axis('equal')

plt.legend(labels=df_for_pie.index, loc='upper right')

plt.show()
```
